# cities_aqi_preprocess

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('cities_20160101-20161231/china_cities_20160101.csv')
df
```

```python
data = df[df['type']=='PM2.5']
data
```

```python
cities = pd.DataFrame(data.columns)
cities = cities.rename(columns={0: "station name"})
```

```python
cities[cities['station name'] == '重庆']
```

```python
cities
```

```python
weathers = pd.read_csv('name-city-english.csv')
```

```python
weathers = weathers.drop(columns=['Unnamed: 0', 'Unknown'])
```

```python
weathers[weathers['station name']=='上海']
```

```python
aqi_weather = pd.merge(cities, weathers, how = 'inner', on = ['station name'])
```

```python
aqi_weather['CityEN']
```

```python
np.isnan(aqi_weather['CityEN'][0])
```

```python
work_df = aqi_weather.copy()
```

```python
work_df
```

```python
work_df = work_df.rename(columns = {'CityEN': 'city'})
```

```python
work_df
```

```python
work_df['city'] = work_df['city'].fillna(work_df['station name'])
```

```python
work_df
```

```python
aqi_weather[aqi_weather['CityEN'].isnull() == True]
```

```python
city_map = {'北京': 'Beijing', '天津':'Tianjin', '上海': 'Shanghai', '重庆':'Chongqing', '库尔勒':'Kuerle', '溧阳':'Liyang'}
```

```python
work_df['city'] = work_df['city'].replace(city_map)
```

```python
work_df
```

```python
work_df.set_index('city', inplace=True)
```

```python
work_df.drop(columns=['Year'], inplace = True)
```

```python
work_df
```

```python
work_df.to_csv('work_station_data.csv')
```
