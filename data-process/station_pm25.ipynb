```python
path='station_PM/station2016/'
```

```python
all_files= os.listdir(path)
```

```python
def getfiles(files):
    li=[]
    for file in files:
        df = pd.read_csv(os.path.join(path,file))
        li.append(df)
    return li
```

```python
frame = getfiles(all_files)
data = pd.concat(frame, axis=0, ignore_index=True)
```

```python
data.head()
```

```python
data_sort = data.sort_values(by='date')
```

```python
df_station = data_sort[data_sort['type']=='PM2.5']
```
