```python
path2 = '../../data/air_data/pm25city/pm25.csv'
```

```python
city25 = pd.read_csv(path2)
```

```python
city25 = city25.drop(['Unnamed: 0'],axis=1)
```

```python
city25.head()
```

```python
city25['date'].dtype
```

```python
city25['hour'].dtype
```

```python
city25['date'] = pd.to_datetime(city25['date'],format='%Y%m%d')
city25.head()
```

```python
city25['hour']= pd.to_timedelta(city25.hour,unit='h')
```

```python
city25.head()
```

```python
city25.loc[:,'Date'] = city25.date+city25.hour
```
